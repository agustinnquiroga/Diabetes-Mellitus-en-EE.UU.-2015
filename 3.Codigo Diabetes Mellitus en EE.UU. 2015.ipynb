**Función que prepara los datos para graficar con porcentajes**

```python
# Esta función recibe un dataframe, una columna del mismo y una lista para categorizar esa columna
# y devuelve un dataframe agrupado por Diabetes y la columna recibida por parámetro más una columna con el porcentaje correspondiente

def df_con_porcentaje_por_columna(df, col, catCol):
    #agrupo estatus de diabetes y col
    df_ret=df[['Diabetes', col]].copy()
    df_ret=df_ret.groupby(['Diabetes', col]).size().reset_index(name = 'Cantidad')
    
    # Ordeno por la columna Diabetes y después por 'col'
    categoria_Diabetes = ['No Diabetes','Prediabetes','Diabetes'] # Lista con el orden establecido
    categoria_col = catCol #Lista con orden recibida por parámetro
    df_ret["Diabetes"] = pd.Categorical(df_ret["Diabetes"], categories = categoria_Diabetes) # Mediante Categorical() se categoriza la columna
    df_ret[col] = pd.Categorical(df_ret[col], categories = categoria_col) # Mediante Categorical() se categoriza la columna
    df_ret.sort_values(["Diabetes",col],inplace=True,ignore_index=True) # Se ordena el DataFrame mediante sort_values(), primero por Diabetes y luego por Frutas, basandose en las listas y categorizadas

    # Ahora calculo el porcentaje de 'col' por el Tipo de Diabetes
    df_ret["Porcentaje"] = df_ret["Cantidad"] # Creamos nueva columna que almacenará luego el porcentaje

    for i in range(0,len(categoria_Diabetes), 1): # recorro el dataframe para cada tipo de diabetes
        for x in range(i*len(categoria_col),(1+i)*len(categoria_col),1): # y para cada tipo de 'col'
            cantidad = df_ret.iloc[x,2] # guardo el valor en la columna cantidad 
            total = df_ret.groupby("Diabetes").sum().Cantidad.values[i] # agrupo por tipo diabetes y guardo la suma de las cantidades 
            df_ret.iloc[x,3] = np.round(cantidad * 100 / total, 2) # calculo el porcentaje y lo guardo en la nueva columna 

    return df_ret
```

Mediante la siguiente función podemos obtener un top de las mayores correlaciones. Veamos pues el top 10.

```python
# Función para obtener las parejas de correlación
def parejas(df):
    '''Obtiene diagonal inferior de parejas en matriz de correlacion'''
    pairs_to_drop = set()  # los conjuntos no permiten duplicados
    cols = df.columns
    for i in range(0, df.shape[1]):
        for j in range(0, i+1):
            pairs_to_drop.add((cols[i], cols[j]))
    return pairs_to_drop

def correlaciones(data, n=5):
    au_corr = data.corr().abs().unstack()
    labels_to_drop = parejas(data)
    au_corr = au_corr.drop(labels=labels_to_drop).sort_values(ascending=False)
    return au_corr[0:n]

print("Correlaciones mas grandes")
print(correlaciones(df,10)) # Saquemos el top 10
```

#### **SelectKBest Feature Selection**


```python
# Separamos matriz de características y variable objetivo
X = df.drop(columns=["Diabetes_012"]) # Matriz de características
y = df.Diabetes_012.astype(int)       # Etiqueta o variable objetivo
```

```python
from imblearn.over_sampling import SMOTE
from collections import Counter
#Balanceo con SMOTE
sm = SMOTE(random_state=42)
X_res, y_res = sm.fit_resample(X, y)

print ("Distribución antes de resampleo {}".format(Counter(y)))
print ("Distribución despues de resampleo {}".format(Counter(y_res)))
```

```python
from sklearn.model_selection import train_test_split # vamos a separar nuestro dataset
# Separamos en entrenamiento/test a razón 70/30
X_train, X_test, y_train, y_test = train_test_split(X_res,y_res,test_size=0.3,random_state=42)
```

```python
# feature selection
from sklearn.feature_selection import mutual_info_classif
def select_features(X_train, y_train, X_test):
	fs = SelectKBest(score_func=mutual_info_classif, k='all') # <--- Acá se define la función mutual_info_classif y evaluamos todas las variables
	fs.fit(X_train, y_train)
	X_train_fs = fs.transform(X_train)
	X_test_fs = fs.transform(X_test)
	return X_train_fs, X_test_fs, fs
```

```python
# llamamos a la función select_features
X_train_fs, X_test_fs, fs = select_features(X_train, y_train, X_test)

# Imprimimos los scores de cada variable
for i in range(len(fs.scores_)):
	print('Feature %d - %s: %f' % (i, X_train.columns[i], fs.scores_[i]))
# Graficamos
plt.bar([i for i in range(len(fs.scores_))], fs.scores_)
plt.show()
```

```python
# Función para visualizar gráficos y métricas de los modelos
def imprimo_resultados (y_test, y_pred):
    fig, axs = plt.subplots(1, 2, figsize=(15, 6))

    # --- Matriz de confusión ---
    cm = confusion_matrix(y_test, y_pred)
    sns.heatmap(cm, annot=True, cmap='crest', fmt="d", cbar=False, ax=axs[0])
    axs[0].set_title("Matriz de confusión")

    # --- Calculamos curva ROC y área para cada clase ---
    # Binarizamos los variables objetivos
    y_test_bin = label_binarize(y_test, classes=[0, 1, 2])
    y_pred_bin = label_binarize(y_pred, classes=[0, 1, 2])
    n_classes = y_pred_bin.shape[1]

    # Calculamos curva ROC y área para cada clase
    fpr = dict()
    tpr = dict()
    roc_auc = dict()
    for i in range(n_classes):
        fpr[i], tpr[i], _ = roc_curve(y_test_bin[:, i], y_pred_bin[:, i])
        roc_auc[i] = auc(fpr[i], tpr[i])
    
    # Graficamos
    lw = 2
    colors = cycle(["aqua", "darkorange", "cornflowerblue"])
    for i, color in zip(range(n_classes), colors):
        axs[1].plot(
            fpr[i],
            tpr[i],
            color=color,
            lw=lw,
            label="Clase {0} (area = {1:0.2f})".format(i, roc_auc[i]),
        )

    axs[1].plot([0, 1], [0, 1], "k--", lw=lw)
    axs[1].set_xlim([0.0, 1.0])
    axs[1].set_ylim([0.0, 1.05])
    axs[1].set_xlabel("Ratio Falsos Positivos")
    axs[1].set_ylabel("Ratio Verdaderos Positivos")
    axs[1].set_title("Curva ROC")
    axs[1].legend(loc="lower right")

    plt.show()
    print()

    # Métricas del modelo
    print(classification_report(y_test, y_pred))
```

```python
def analizo_modelo(modelo):
    # Ajuste
    modelo.fit(X_train_fs, y_train)

    # Predicciones
    y_pred = modelo.predict(X_test_fs)

    # Muestro las métricas
    imprimo_resultados(y_test, y_pred)
```

```python
#Dataframes para guardar resultados de calculos de métricas
accuracy   = pd.DataFrame(columns=["name", "accuracy"])
precision = pd.DataFrame(columns=["name", "precision"])
recall = pd.DataFrame(columns=["name", "recall"])
f1_score_ = pd.DataFrame(columns=["name", "score"])
tiempo = pd.DataFrame(columns=["name", "time"])
```

General

```python
#Calculo las métricas para cada modelo
for modelo in modelos:
    inicio=time.time()
    y_pred = modelo['model'].predict(X_test) # Predicción
    fin = time.time()
    intervalo = fin-inicio
    
    # 1. Accuracy
    model_acc = {"name": modelo["name"], "accuracy": accuracy_score(y_test, y_pred)}
    accuracy = accuracy.append(model_acc, ignore_index=True)

    # 2. Precision
    model_pre = {"name": modelo["name"], "precision": precision_score(y_test, y_pred, average='macro')}
    precision = precision.append(model_pre, ignore_index=True)
    
    # 3. Recall
    model_rec = {"name": modelo["name"], "recall": recall_score(y_test, y_pred, average='macro')}
    recall = recall.append(model_rec, ignore_index=True)
    
    # 4. F1 score
    model_f1_score = {"name": modelo["name"], "score": f1_score(y_test, y_pred, average='macro')}
    f1_score_ = f1_score_.append(model_f1_score, ignore_index=True)

    #5. Tiempo
    model_time = {"name": modelo["name"], "time": intervalo}
    tiempo = tiempo.append(model_time, ignore_index=True)

#Imprimo las métricas en orden descendente
print(" Accuracy ".center(80, "="))
print(accuracy.sort_values(by="accuracy", ascending=False))
print()
print(" Precision ".center(80, "="))
print(precision.sort_values(by="precision", ascending=False))
print()
print(" Recall ".center(80, "="))
print(recall.sort_values(by="recall", ascending=False))
print()
print(" F1 Score ".center(80, "="))
print(f1_score_.sort_values(by="score", ascending=False))
print()
print(" Tiempo de inferencia ".center(80, "="))
print(tiempo.sort_values(by="time", ascending=True))
print()
```

También se observa que el segundo algoritmo mejor posicionado es el Bagging Classifier y que incluso tiene un menor tiempo de inferencia, siendo 9 veces más rápido que el Random Forest. Por otra parte, se observa que KNN es el tercero mejor posicionado en cuanto a métricas, fue muy rápido en tiempo de entrenamiento pero extremadamente lento en predicción.
